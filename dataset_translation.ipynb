```python
from datasets import load_dataset
from deep_translator import GoogleTranslator
import pandas as pd
from tqdm import tqdm
```

```python
# Load dataset
dataset = load_dataset("linagora/linto-dataset-audio-ar-tn", split="train")
df = dataset.to_pandas()
```

```python
# Translate transcripts
translator = GoogleTranslator(source='ar', target='fr')

# Use tqdm for progress bar
tqdm.pandas()
```

```python
def safe_translate(text):
    from deep_translator.exceptions import TranslationNotFound
    try:
        return translator.translate(text)
    except TranslationNotFound:
        return "NoTranslation"
    except Exception as e:
        print(f"Error translating '{text[:30]}...': {e}")
        return "NoTranslation"

# Apply translation with progress bar
tqdm.pandas()
df['transcript_fr'] = df['transcript'].progress_apply(safe_translate)

# Save to CSV
df.to_csv("translated_dataset_fr.csv", index=False)

# Show sample
print(df[['transcript', 'transcript_fr']].head())
```

```python
from deep_translator import GoogleTranslator
from tqdm import tqdm

translator = GoogleTranslator(source='ar', target='fr')

tqdm.pandas()

def safe_translate(text):
    from deep_translator.exceptions import TranslationNotFound
    try:
        return translator.translate(text)
    except TranslationNotFound:
        return "NoTranslation"
    except Exception as e:
        print(f"Error translating '{text[:30]}...': {e}")
        return "NoTranslation"
```

```python
import pandas as pd

# Read in chunks
chunk_size = 1000  # Adjust based on memory
for chunk in pd.read_csv('translated_dataset_fr.csv', chunksize=chunk_size):
    # Check if any rows in the chunk have 'transcript_fr' == 'NoTranslation'
    if (chunk['transcript_fr'] == 'NoTranslation').any():
        # Apply safe_translate only to rows where transcript_fr is 'NoTranslation'
        chunk.loc[chunk['transcript_fr'] == 'NoTranslation', 'transcript_fr'] = (
            chunk.loc[chunk['transcript_fr'] == 'NoTranslation', 'transcript_fr'].apply(safe_translate)
        )
    print('Chunk processed')
```

```python
import pandas as pd
import os
import shutil

# Define input and temporary output file paths
input_file = 'translated_dataset_fr.csv'
temp_output_file = 'translated_dataset_fr_temp.csv'

chunk_size = 5000  
for i, chunk in enumerate(pd.read_csv(input_file, chunksize=chunk_size)):
    if (chunk['transcript_fr'] == 'NoTranslation').any():
        chunk.loc[chunk['transcript_fr'] == 'NoTranslation', 'transcript_fr'] = (
            chunk.loc[chunk['transcript_fr'] == 'NoTranslation', 'transcript_fr'].apply(safe_translate)
        )
    
    mode = 'w' if i == 0 else 'a'
    header = i == 0
    chunk.to_csv(temp_output_file, mode=mode, header=header, index=False)
    print(f'Chunk {i+1} processed')

shutil.move(temp_output_file, input_file)
print(f'Original file {input_file} updated with translations')
```

```python
import pandas as pd
import os

# Define input file and output directory
input_file = 'translated_dataset_fr.csv'
output_dir = 'translation_results'

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Read in chunks
chunk_size = 2000  # Adjust based on memory
for i, chunk in enumerate(pd.read_csv(input_file, chunksize=chunk_size)):
    # Define output file path for this chunk
    output_file = os.path.join(output_dir, f'chunk_{i+1}.txt')
    
    # Write chunk inspection to the text file
    with open(output_file, 'w', encoding='utf-8') as f:
        for _, row in chunk.iterrows():
            f.write(f"{row['transcript']}\n{row['transcript_fr']}\n")
            f.write('\n\n' + '---' * 20 + '\n')
    
    # Optional: Print a confirmation to terminal
    print(f'Chunk {i+1} saved to {output_file}')
```
